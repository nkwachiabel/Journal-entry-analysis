```python
import pandas as pd
journal_entries = pd.read_excel("Vinatics 2020 Journal Entries.xlsx")
journal_entries.head()
```

```python
weekend_journal_entries = journal_entries.copy()
weekend_journal_entries['Entry Date'] = pd.to_datetime(weekend_journal_entries['Entry Date'])
weekend_journal_entries['Weekday Number'] = weekend_journal_entries['Entry Date'].dt.dayofweek
weekend_journal_entries.head()
```

```python
weekend_journal_entries = weekend_journal_entries[weekend_journal_entries['Weekday Number'] > 4]
weekend_journal_entries.head()
```

```python
journal_entries_all_employees = journal_entries.copy()
journal_entries_all_employees = journal_entries_all_employees.pivot_table(index=['JE number','Preparer ID'])
journal_entries_all_employees = pd.DataFrame(journal_entries_all_employees.to_records())
journal_entries_all_employees.head()
```

```python
journal_entries_all_employees_count = journal_entries_all_employees['Preparer ID'].value_counts()
journal_entries_all_employees_count
```

```python
employee_name = 'Codi Padgett'
journal_entries_by_employee = journal_entries.copy()
journal_entries_by_employee = journal_entries_by_employee[journal_entries_by_employee['Preparer ID'] == employee_name]
journal_entries_by_employee
```

```python
post_dated_journals = journal_entries.copy()
post_dated_journals = post_dated_journals[['JE number','GL Code','Period','Entry Date','Effective Date','Functional Amount']]
post_dated_journals['Date Difference'] = post_dated_journals['Effective Date'] - post_dated_journals['Entry Date']
post_dated_journals.sort_values(by = ['Date Difference'])
```
